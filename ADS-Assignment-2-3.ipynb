```python
#Import packages to use

import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Load Dataset

df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
df.head()
```

```python
#Data types

df.dtypes
```

```python
#Data info

df.info()
```

```python
# View distance from home by job role and attrition

df.groupby(['JobRole', 'Attrition'])['DistanceFromHome'].mean().unstack()
```

```python
# Comparison of average monthly income by education and attrition.

df.groupby(['Education','Attrition'])['MonthlyIncome'].mean().unstack()
```

```python
# Visualize distance from home by job role and attrition

fig, axes = plt.subplots(1, figsize=(13,9))
fig.suptitle('Distance from home by job role and attrition')

sn.barplot(ax=axes, x='JobRole', y='DistanceFromHome', hue='Attrition', data=df)
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
# Visualize comparison of average monthly income by education and attrition.

fig, axes = plt.subplots(1, figsize=(8,9))
fig.suptitle('Comparison of average monthly income by education and attrition')
sn.barplot(ax=axes, x='Education', y='MonthlyIncome', hue='Attrition', data=df)
```
